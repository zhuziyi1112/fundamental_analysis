## Sector Historical Performance

```python
# API URL and request here (as before)
url_historical_sectors_performance = f"https://financialmodelingprep.com/api/v3/historical-sectors-performance?from=2022-01-01&to=2024-12-31&apikey={api_key}"

try:
    historical_sectors_performance = requests.get(url_historical_sectors_performance)
    historical_sectors_performance.raise_for_status()  # Raise an error for bad responses
    historical_sectors_performance = historical_sectors_performance.json()  # Parse JSON response
except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
    historical_sectors_performance = []

# Create DataFrame for Historical Sectors Performance
if not historical_sectors_performance:
    print("No data received from the API.")
    df_historical_sectors_performance = pd.DataFrame()  # Create an empty DataFrame
else:
    df_historical_sectors_performance = pd.DataFrame(historical_sectors_performance)

    # Convert 'date' to datetime and set as index
    df_historical_sectors_performance['date'] = pd.to_datetime(df_historical_sectors_performance['date'])
    df = df_historical_sectors_performance.sort_values(by='date', ascending=True).reset_index(drop=True)
    df.set_index('date', inplace=True)

    # List of rolling periods
    rolling_periods = [5, 20]
    newly_added_columns = []  # List to track newly added columns

    # Initialize a new DataFrame to store rolling changes
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            # Calculate growth factors and cumulative growth
            growth_factors = 1 + (df[column] / 100)
            cumulative_growth = growth_factors.cumprod()

            for period in rolling_periods:
                # Calculate rolling change for the specified period
                rolling_change = ((cumulative_growth / cumulative_growth.shift(period)) - 1) * 100
                new_column_name = f'{period}-Day Rolling Change % - {column}'
                df[new_column_name] = rolling_change.fillna(0)  # Filling with 0
                newly_added_columns.append(new_column_name)  # Track the new column

    # Select only the newly added columns
    df_historical_sectors_performance = df.sort_values(by='date', ascending=False).reset_index()
    df_new_columns = df_historical_sectors_performance[['date'] + newly_added_columns]

    # Prepare to melt the DataFrame
    df_new_columns.reset_index(drop=True, inplace=True)

    # Use pd.melt to convert the DataFrame to long format
    df_melted = pd.melt(df_new_columns, id_vars=['date'], var_name='rolling_change_type', value_name='percentage_change')

    # Split the 'rolling_change_type' into two new columns
    df_melted[['rolling_period', 'sector_name']] = df_melted['rolling_change_type'].str.split(' - ', expand=True)

# Example output of the final melted and split DataFrame
df_historical_sectors_performance = df_melted[['date','sector_name', 'rolling_period', 'percentage_change']]

# Save the combined DataFrame to a CSV file
subdirectory = 'data_outputs'  # This is your target subfolder
file_name = 'historical_sectors_performance.csv'
file_path = os.path.join(current_directory, subdirectory, file_name)  # Full path for the file

df_historical_sectors_performance.to_csv(file_path, index=False)
```

```python
import requests
import pandas as pd
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from transformers import pipeline

# Initialize VADER sentiment analyzer
vader_analyzer = SentimentIntensityAnalyzer()

# Initialize TextBlob sentiment analysis function
def textblob_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Initialize transformers sentiment analysis pipeline
transformer_sentiment = pipeline("sentiment-analysis")

# URL to fetch the latest articles
url_latest_articles = f"https://financialmodelingprep.com/api/v3/fmp/articles?page=0&size=10000&apikey={api_key}"

try:
    # Sending request to the URL
    latest_articles = requests.get(url_latest_articles)
    latest_articles.raise_for_status()  # Raises an error for bad responses
    
    # Parse JSON response
    latest_articles = latest_articles.json()  
    
    # Create DataFrame from the JSON response
    df_latest_articles = pd.DataFrame(latest_articles['content'])  # Adjust based on the actual API response structure
    
    # Check for existence of 'tickers' column and split to extract tickers
    if 'tickers' in df_latest_articles.columns:
        df_latest_articles['tickers'] = df_latest_articles['tickers'].str.split(':').str[1]
    
    # Generate sentiment scores
    df_latest_articles['vader_sentiment_score'] = df_latest_articles['content'].apply(lambda x: vader_analyzer.polarity_scores(x)['compound'])
    df_latest_articles['textblob_sentiment_score'] = df_latest_articles['content'].apply(textblob_sentiment)

    # Generate sentiment score using the transformers pipeline, handling long texts
    def safe_transformer_sentiment(text):
        # Split long text into manageable chunks (at most 512 characters)
        if len(text) > 512:
            text = text[:512]  # Truncate to 512 characters for this example
        result = transformer_sentiment(text)
        return result[0]['score'] if result[0]['label'] == 'POSITIVE' else -result[0]['score']

    df_latest_articles['transformer_sentiment_score'] = df_latest_articles['content'].apply(safe_transformer_sentiment)

    # Calculate average sentiment score
    df_latest_articles['average_sentiment_score'] = df_latest_articles[['vader_sentiment_score', 'textblob_sentiment_score', 'transformer_sentiment_score']].mean(axis=1)

    # Select relevant columns
    df_latest_articles = df_latest_articles[['date', 'tickers', 'content', 'author', 'vader_sentiment_score', 'textblob_sentiment_score', 'transformer_sentiment_score', 'average_sentiment_score']]
    
    # Display the resulting DataFrame
    display(df_latest_articles)
    
except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
```
